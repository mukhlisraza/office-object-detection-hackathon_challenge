```python
import os
import torch
import wandb
from wandb.integration.ultralytics import add_wandb_callback
from ultralytics import YOLO
from IPython.display import display, Image
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
torch.cuda.empty_cache()
device
```

```python
wandb.login()
```

```python
# Sweep configuration
sweep_configuration = {
    "name": "smart-office-final-sweep",
    "method": "bayes",
    "metric": {
        "name": "metrics/mAP50-95(B)",
        "goal": "maximize"
    },
    "parameters": {
        "model": {
            "values": ['yolov8n', 'yolov8s', 'yolov8m', 'yolov8l', 'yolov8x']
        },
        "pretrained": {
            "values": [True, False]
        },
        "lr0": {
            "min": 0.0001, 
            "max": 0.01
        },
        "batch_size": {
            "values": [4, 8, 12, 16]
        },
        "optimizer": {
            "values": ["AdamW", "SGD", "Adam", "RMSProp"]
        },
        "imgsz": {
            "values": [640]
        },
        "epochs": {
            "values": [50, 100, 150, 200]
        },
        "val": {
            "values": [True]
        },
        "plots": {
            "values": [True]
        },
        "cos_lr": {
            "values": [True, False]
        },
        "amp": {
            "values": [True, False]
        },
        "single_cls": {
            "values": [False]
        },

    }
}
```

```python
sweep_id = wandb.sweep(sweep=sweep_configuration, project="smart-office")
```

```python
sweep_id
```

```python
def train():
    with wandb.init() as run:
        config = run.config
        model = YOLO(f"{config.model}.pt").to('cuda' if torch.cuda.is_available() else 'cpu')
            
        try:
            # Start training with the given configuration
            results = model.train(
                project="office-objects-detections",
                data=f'../data/data.yaml',
                epochs=config.epochs,
                imgsz=config.imgsz,
                batch=config.batch_size,
                lr0=config.lr0,
                workers=8,
                plots=config.plots,
                pretrained=config.pretrained,
                device='cuda' if torch.cuda.is_available() else 'cpu',
                optimizer=config.optimizer,
                name='yolos-updated-dataset-trail2'
            )
        
            # Log the results
            wandb.log({"loss": results})
        
        except Exception as e:
            print(f"An error occurred: {e}")

# Use the existing sweep ID to start the sweep
wandb.agent(sweep_id, function=train, count=5)
```

# Testing

## Testing script run on whole directory

```python
import os
from ultralytics import YOLO
from IPython.display import display, Image

model = YOLO('office-objects-detections/yolos11/weights/best.pt')
test_dir = 'test-images'
results = model.predict(source=test_dir, plots=True, save=True, conf=0.5)

for result in results:
    annotated_image_path = result.save_dir + f'/{os.path.basename(result.path)}'  # Match original filename
    if os.path.exists(annotated_image_path):
        display(Image(filename=annotated_image_path))
    else:
        print(f"Annotated image not found at {annotated_image_path}")
```
